## Automatic Differentiation
PyTorch can automatically compute gradients.

```
# Create tensor with gradient tracking
x = torch.tensor([2.0], requires_grad=True)

# Forward pass
y = x ** 2  # y = x^2
z = 2 * y + 3  # z = 2x^2 + 3

# Backward pass
z.backward()

# Check gradient: dz/dx = 4x = 8
print(f'x = {x.item()}')
print(f'z = {z.item()}')
print(f'dz/dx = {x.grad.item()}')  # Should be 8
```

## Linear Regression Example

```
# Parameters
w = torch.tensor([1.0], requires_grad=True)
b = torch.tensor([0.0], requires_grad=True)

# Data
x = torch.tensor([1.0, 2.0, 3.0, 4.0])
y_true = torch.tensor([2.0, 4.0, 6.0, 8.0])  # y = 2x

# Training loop
lr = 0.1
for epoch in range(100):
    # Forward
    y_pred = w * x + b
    loss = ((y_pred - y_true) ** 2).mean()
    
    # Backward
    loss.backward()
    
    # Update
    with torch.no_grad():
        w -= lr * w.grad
        b -= lr * b.grad
        w.grad.zero_()
        b.grad.zero_()

print(f'Learned: y = {w.item():.2f}x + {b.item():.2f}')
print(f'Expected: y = 2x + 0')
```
